## 1. ¿Qué es GitHub Pages?

**GitHub Pages** es un servicio gratuito de GitHub que permite publicar sitios web directamente desde un repositorio. No necesitas servidor propio, ni pagar hosting: GitHub se encarga de todo.

### Características principales

- ✅ **Gratuito** para repositorios públicos.
- ✅ **HTTPS** automático (certificado SSL incluido).
- ✅ **Dominio gratuito** con el patrón `https://<usuario>.github.io/<repositorio>`.
- ✅ **Integración continua**: cada `git push` actualiza el sitio automáticamente.
- ✅ Compatible con **Jekyll**, **React** (build estático), **Vue**, **Hugo**, etc.

### Tipos de sitios en GitHub Pages

| Tipo | URL | Repositorio |
|------|-----|-------------|
| Usuario / Organización | `https://usuario.github.io` | `usuario/usuario.github.io` |
| Proyecto | `https://usuario.github.io/proyecto` | Cualquier repo |

Para un sitio académico personal o de grupo de investigación, lo más común es el **sitio de usuario** o el **sitio de proyecto**.

## 7. JavaScript — `js/main.js`

Un poco de JavaScript para añadir interactividad: filtro de publicaciones por año y menú hamburguesa en móvil.

````python
main_js = '''
// ===================================================
// Filtro de publicaciones por año
// ===================================================
document.querySelectorAll('.filter-btn').forEach(btn => {
  btn.addEventListener('click', () => {
    const year = btn.dataset.year;

    // Marcar botón activo
    document.querySelectorAll('.filter-btn').forEach(b => b.classList.remove('active'));
    btn.classList.add('active');

    // Mostrar / ocultar publicaciones
    document.querySelectorAll('.pub-item').forEach(item => {
      if (year === 'all' || item.dataset.year === year) {
        item.style.display = 'flex';
      } else {
        item.style.display = 'none';
      }
    });
  });
});

// ===================================================
// Menú hamburguesa (móvil)
// ===================================================
const hamburger = document.querySelector('.hamburger');
const navLinks  = document.querySelector('.nav-links');

if (hamburger) {
  hamburger.addEventListener('click', () => {
    navLinks.classList.toggle('open');
    hamburger.setAttribute(
      'aria-expanded',
      hamburger.getAttribute('aria-expanded') === 'false' ? 'true' : 'false'
    );
  });
}

// ===================================================
// Scroll suave para los anclas internas
// ===================================================
document.querySelectorAll('a[href^="#"]').forEach(anchor => {
  anchor.addEventListener('click', e => {
    e.preventDefault();
    const target = document.querySelector(anchor.getAttribute('href'));
    if (target) target.scrollIntoView({ behavior: 'smooth', block: 'start' });
  });
});
'''

print("Código JavaScript listo.")
````

## 10. Opción avanzada: usar Jekyll

[Jekyll](https://jekyllrb.com/) es el generador de sitios estáticos que GitHub Pages usa internamente. Con Jekyll puedes:

- Escribir tu contenido en **Markdown** (mucho más fácil que HTML).
- Usar **plantillas** (layouts) para no repetir el navbar y el footer en cada página.
- Administrar un **blog/noticias** automáticamente.
- Cambiar todo el diseño cambiando un solo **tema**.

### 10.1 Estructura de un proyecto Jekyll

```
observatorio-ova/
├── _config.yml          ← Configuración global
├── _layouts/
│   ├── default.html     ← Layout base con navbar y footer
│   └── post.html        ← Layout para noticias/posts
├── _includes/
│   ├── navbar.html
│   └── footer.html
├── _posts/              ← Noticias en Markdown
│   ├── 2026-03-10-articulo-apj.md
│   └── 2026-01-20-premio-luna.md
├── _data/
│   ├── equipo.yml       ← Datos del equipo en YAML
│   └── publicaciones.yml
├── assets/
│   ├── css/estilos.css
│   └── img/
├── index.md
├── investigacion.md
├── equipo.md
└── publicaciones.md
```

### 10.2 Ejemplo: `_config.yml`

````python
config_yml = """
# _config.yml — Configuración de Jekyll para el OVA

title:       "Observatorio Virtual de Astronomía"
description: "Investigación en cosmología, exoplanetas y astrofísica de alta energía."
url:         "https://TU_USUARIO.github.io"
baseurl:     "/observatorio-ova"

author:
  name:  "Equipo OVA"
  email: "ova@astronomia.cl"

# Tema (puedes cambiar por 'minima', 'minimal-mistakes', 'just-the-docs', etc.)
theme: minima

# Plugins
plugins:
  - jekyll-feed        # RSS automático
  - jekyll-seo-tag     # Meta tags SEO automáticos
  - jekyll-sitemap     # sitemap.xml automático

# Colecciones personalizadas
collections:
  publicaciones:
    output: true
    permalink: /publicaciones/:name/

# Excluir del build
exclude:
  - Gemfile
  - Gemfile.lock
  - README.md
"""

print("Configuración _config.yml de Jekyll:")
print(config_yml)
````

## 12. Buenas prácticas

### 12.1 Accesibilidad (a11y)

| Práctica | Ejemplo |
|----------|---------|
| `alt` en todas las imágenes | `<img src="nebula.jpg" alt="Nebulosa de Orión en infrarrojo">` |
| `aria-label` en íconos | `<a aria-label="Twitter de OVA">` |
| Contraste de color suficiente | Fondo `#0B1120` + texto `#F5F5F5` → ratio > 10:1 ✅ |
| Navegación por teclado | No usar `outline: none` sin alternativa visible |

### 12.2 SEO (posicionamiento en buscadores)

```html
<!-- En el <head> de cada página -->
<meta name="description" content="Grupo de investigación en cosmología y exoplanetas." />
<meta property="og:title"       content="OVA — Observatorio Virtual de Astronomía" />
<meta property="og:description" content="Investigación en cosmología y exoplanetas." />
<meta property="og:image"       content="https://usuario.github.io/observatorio-ova/img/og-card.jpg" />
<meta name="twitter:card"       content="summary_large_image" />
```

### 12.3 Rendimiento

- 📦 **Comprimir imágenes** antes de subir: usa [Squoosh](https://squoosh.app/) o `ImageMagick`.
- 🔠 **Cargar fuentes solo en pesos necesarios**: `Orbitron:wght@400;700` no `:wght@100..900`.
- ♻️ **Cachear recursos**: GitHub Pages ya configura headers de cache automáticamente.
- 🏎️ **Lazy-loading de imágenes**: `<img loading="lazy" ...>`.

### 12.4 Estructura de commits recomendada (Conventional Commits)

```bash
feat:     nueva página de galería fotográfica
fix:      corregir enlace roto en publicaciones.html
docs:     actualizar README con instrucciones de contribución
style:    ajustar padding de tarjetas en móvil
content:  agregar artículo ApJ vol. 950
```
